```python
# Create and plot bulge and disk template, for Nb=528, Nd=2102 - my best fit values
Nbulge = 528
Ndisk = 2102
alpha=2.6
n=2.35
sigma=1.528
z0=0.7
beta=1.2

import bulge_pscounts as bps
import disk_pscounts as dps

btemp = np.zeros((12,12,8))
dtemp = np.zeros((12,12,8))

for i in range(12):
    for j in range(12):
        for k in range(8):
            omega_val = omega_ijk[j,k]
            btemp[i,j,k] = bps.Nbulge_ijk(i, j, k, Nbulge, omega_val, alpha, beta)
            dtemp[i,j,k] = dps.Ndisk_ijk(i, j, k, Ndisk, omega_val, n, sigma, z0, beta)
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
# transpose as it plots the first index on the y-axis
cax = ax.matshow(np.transpose(np.sum(btemp,axis=2)), interpolation='nearest')
fig.colorbar(cax)
plt.show()
                 
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
# transpose as it plots the first index on the y-axis
cax = ax.matshow(np.transpose(np.sum(dtemp,axis=2)), interpolation='nearest')
fig.colorbar(cax)
plt.show()

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
# transpose as it plots the first index on the y-axis
cax = ax.matshow(np.transpose(np.sum(dtemp+btemp,axis=2)), interpolation='nearest')
fig.colorbar(cax)
plt.show()
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
# transpose as it plots the first index on the y-axis
cax = ax.matshow(np.transpose(btemp[:,:,0]), interpolation='nearest')
fig.colorbar(cax)
plt.show()
                 
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
# transpose as it plots the first index on the y-axis
cax = ax.matshow(np.transpose(dtemp[:,:,0]), interpolation='nearest')
fig.colorbar(cax)
plt.show()

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
# transpose as it plots the first index on the y-axis
cax = ax.matshow(np.transpose(btemp[:,:,0]+dtemp[:,:,0]), interpolation='nearest')
fig.colorbar(cax)
plt.show()
```
